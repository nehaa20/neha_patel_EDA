```python
#analysing customer bought which item
merged_customer_data = pd.merge(transaction_data, customer_data, on = 'CustomerID', how = "inner")
```

```python
merged_customer_data.info()
```

```python
merged_products_data = pd.merge(products_data,merged_customer_data, on = 'ProductID', how= "inner")
```

```python
merged_products_data.info()
```

```python
merged_products_data.head(10)
```

```python
merged_products_data = merged_products_data[['ProductID', 'ProductName', 'Category', 'Price_x', 'TransactionID', 'CustomerID', 'TransactionDate', 'Quantity', 'TotalValue', 'CustomerName', 'Region', 'SignupDate']]
```

```python
merged_products_data.info()
```

```python
merged_products_data.head(10)
```

```python
# Analyzing different types of products names 
unique_products = merged_products_data.groupby('ProductID')['ProductName'].unique()
print(unique_products)
```

```python
#Analyzing group of people bought same product
customer_name = merged_products_data.groupby('ProductID')['CustomerName'].unique()
print(customer_name)
```

```python
#Analyzing what product did customer bought
product_name_mapping = merged_products_data.drop_duplicates(subset=['ProductID'])[['ProductID', 'ProductName', 'CustomerName']].set_index('ProductID')
print("\nProduct names mapped to Product IDs:\n", product_name_mapping)
```

```python
#Analyzing quantity of product bought by the customer
product_name_mapping = merged_products_data.drop_duplicates(subset=['ProductID'])[['ProductID', 'ProductName', 'CustomerName', 'Quantity']].set_index('ProductID')
print("\nProduct names mapped to Product IDs:\n", product_name_mapping)
```

```python
#analyzing the trasaction date
product_name_mapping = merged_products_data.drop_duplicates(subset=['ProductID'])[['ProductID', 'ProductName', 'CustomerName', 'Quantity', 'TransactionDate']].set_index('ProductID')
print("\nProduct names mapped to Product IDs:\n", product_name_mapping)
```
